```python
import pandas as pd
import numpy as np
import json

triangles = pd.read_json("../data/triangles.json")

dual_chart = pd.read_csv("../data/dual-chart.csv")
dual_chart.set_index("Attacking", inplace = True)

triangles
```

```python
def stability_calculator(A, B, C, W):

    offense = [
        W.loc[A,B],
        W.loc[B,C],
        W.loc[C,A]
    ]

    defense = [
        W.loc[A,C],
        W.loc[C,B],
        W.loc[B,A]
    ]

    return np.var(offense) + np.var(defense)
```

```python
triangles["score"] = triangles.apply(
    lambda row : stability_calculator(row[0], row[1], row[2], dual_chart),
    axis = 1
)

triangles = triangles.sort_values(by = "score", ascending = True)
```
